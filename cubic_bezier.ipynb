```python
# homework 5. Implement Cubic Bezier

# 4 control points p0,p1,p2,p3 are specified and points along the Bezier curve are 
# computed by varying the parameter t from 0 to 1. numpy.linspace is used to generate 
# evenly spaced values of t between 0 and 1, and then corresponding points on the curve 
# are computed using the cubic_bezier function. Finally, we print out the computed points.

import numpy as np 

def cubic_bezier(t, p0, p1, p2, p3):
    """
    Compute a point on a cubic Bezier curve.

    Parameters:
    t (float): Parameter ranging from 0 to 1.
    p0, p1, p2, p3 (tuple): Control points (x, y).

    Returns:
    tuple: Coordinates of the point on the Bezier curve.
    """
    x = (1 - t) ** 3 * p0[0] + 3 * (1 - t) ** 2 * t * p1[0] + 3 * (1 - t) * t ** 2 * p2[0] + t ** 3 * p3[0]
    y = (1 - t) ** 3 * p0[1] + 3 * (1 - t) ** 2 * t * p1[1] + 3 * (1 - t) * t ** 2 * p2[1] + t ** 3 * p3[1]
    return x, y

# control points
p0 = (0, 0)
p1 = (1, 2)
p2 = (3, 1)
p3 = (4, 3)

# Compute points along the curve
num_points = 100
points = [cubic_bezier(t, p0, p1, p2, p3) for t in np.linspace(0, 1, num_points)]

# Print the points
for point in points:
    print(point)
```
